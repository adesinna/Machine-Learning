```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,21)
y =[]
for i in x:
    calc = pow(i,2)
    y.append(calc)

plt.plot(x,y, label='y^2', color='red', linewidth=2,linestyle='--', marker='.', markersize=10)

plt.title('My first graph', fontdict={'fontname':'Courier New', 'fontsize':20 })
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.xticks(np.arange(0,21,2))
plt.yticks(np.arange(0,501,50))

plt.legend(loc = 'upper left', bbox_to_anchor = (1,1))
plt.rcParams['figure.figsize'] = 10,5

plt.show()
```

```python
x = np.arange(0,21)
y = []
z = []
for i in x:
    calc_y = pow(i,2)
    calc_z = (2*i) + 4
    y.append(calc_y)
    z.append(calc_z)



plt.plot(x,y, label='y^2', color='red', linewidth=2,linestyle='-', marker='.', markersize=5)
plt.plot(x,z, label='2x + 4', color='blue', linewidth=2,linestyle='-', marker='o', markersize=5)


plt.title('My first graph', fontdict={'fontname':'Courier New', 'fontsize':20 })
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.xticks(np.arange(0,21,2))
plt.yticks(np.arange(0,501,50))


plt.legend(loc = 'upper left', bbox_to_anchor = (1,1))


plt.rcParams['figure.figsize'] = 10,5
plt.show()
```

```python
# Stripping lbs
fifa_data.Weight
```

```python
fifa_data.Weight = [float(i.strip('lbs')) if type(i)==str else i for i in fifa_data.Weight] # List comprehension
```

```python
fifa_data.Weight.describe()
```

```python
light = fifa_data.loc[(fifa_data.Weight < 125)].count()[0]
light_medium = fifa_data.loc[(fifa_data.Weight >= 125) & (fifa_data.Weight < 150)].count()[0]
medium = fifa_data.loc[(fifa_data.Weight >= 150) & (fifa_data.Weight < 175)].count()[0]
medium_heavy = fifa_data.loc[(fifa_data.Weight >= 175) & (fifa_data.Weight < 200)].count()[0]
heavy = fifa_data.loc[(fifa_data.Weight >= 200)].count()[0]
```

```python
weights = [light, light_medium, medium, medium_heavy, heavy]
labels =  ['light', 'light_medium', 'medium', 'medium_heavy', 'heavy']
explode= (.4, .2, 0, 0, .4) # in precentage so must add to 1

plt.style.use('ggplot') # for nice styling


plt.pie(weights,labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)
plt.title('Weights of players', fontdict={'fontname':'Courier New', 'fontsize':20 })

plt.rcParams['figure.figsize'] = 20,10
plt.show()
```

### Plot Box

See box image for more explanation

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa_data.loc[fifa_data.Club == "FC Barcelona"]['Overall']
madrid = fifa_data.loc[fifa_data.Club == "Real Madrid"]['Overall']
revs = fifa_data.loc[fifa_data.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
