```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-10, 11)
y =[]
for i in x:
    calc = pow(i,2) + i - 2
    y.append(calc)

plt.plot(x,y, label='x^4', color='red', linewidth=2,linestyle='--', marker='.', markersize=10)
plt.title('My first graph', fontdict={'fontname':'Courier New', 'fontsize':20 })
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.axhline(y=0, color='blue', linestyle='-')



plt.legend(loc = 'upper left', bbox_to_anchor = (1,1))
plt.rcParams['figure.figsize'] = 10,5

plt.show()
```

```python
x = np.arange(-20, 21)
y =[]
for i in x:
    calc = np.absolute(i) + np.absolute(i - 4)
    y.append(calc)

plt.plot(x,y, label='|x| + |x-4|', color='red', linewidth=2,linestyle='--', marker='.', markersize=10)
plt.title('My first graph', fontdict={'fontname':'Courier New', 'fontsize':20 })
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.axhline(y=0, color='blue', linestyle='-')



plt.legend(loc = 'upper left', bbox_to_anchor = (1,1))
plt.rcParams['figure.figsize'] = 10,5

plt.show()
```

```python

x = np.arange(-20, 21)
y =[]
for i in x:
    calc = pow(i,2) - (2*i) + 1
    y.append(calc)

plt.plot(x,y, label='x^2 - 2x + 1', color='red', linewidth=2,linestyle='--', marker='.', markersize=10)
plt.title('My first graph', fontdict={'fontname':'Courier New', 'fontsize':20 })
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.axhline(y=0, color='blue', linestyle='-')



plt.legend(loc = 'upper left', bbox_to_anchor = (1,1))
plt.rcParams['figure.figsize'] = 10,5

plt.show()
```
